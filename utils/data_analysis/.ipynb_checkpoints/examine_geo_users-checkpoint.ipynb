# Loading User Locations

```python
import pandas as pd
import common.constants as c
from load_locations import load_locations_df
```

```python
def load_discord_df(gdf_mode):
    '''
    Dataframe to show how much variation there is:
    '''

    gdf_discord = gdf_mode.groupby(gdf_mode[c.GEO_DEV]).agg({
          c.U_ID: 'count',
          c.NUM_TWEETS: 'mean',
        })
    gdf_discord.reset_index(inplace=True)
    return gdf_discord
```

```python
gdf_mode = load_locations_df()
gdf_mode = gdf_mode.set_crs(epsg=2163)  # Convert flat USA coords
gdf_mode = gdf_mode.to_crs(epsg=4326)   # to lat/long coords
gdf_mode
```

```python
perc_range = [0.1*x for x in range(1,10)]
gdf_mode.describe(percentiles=perc_range)
```

```python
import matplotlib.pyplot as plt

# First, create 'gdf_discord' based on number of users with each distance away.
gdf_discord = gdf_mode.groupby(gdf_mode[c.GEO_DEV]).agg({c.U_ID: 'count'})
gdf_discord.reset_index(inplace=True)

# Then, round all distances to nearest 100s.
gdf_discord[c.GEO_DEV] = gdf_discord[c.GEO_DEV].round(decimals=-2)
gdf_discord = gdf_discord.groupby(gdf_discord[c.GEO_DEV]).agg({c.U_ID: 'count'})
gdf_discord.reset_index(inplace=True)

print(gdf_discord)

# Finally, plot it:
plt.plot(gdf_discord[c.GEO_DEV], gdf_discord[c.U_ID])
```

```python
import folium
from folium.plugins import HeatMap, HeatMapWithTime
```

```python
def save_heatmap(points, filepath):
    '''
    Create a Folium heatmap of a set of points.
    Expects a list of point objects with an x and y value, or a list of lists of point objects (where each list = timestep).
    
    ** MAY NOT WORK RN.
    '''

    m = folium.Map(
        location=[-82.44269374999999, 27.9965945],
        tiles="cartodbpositron",
        zoom_start=2,
    )

    if isinstance(points[0], list):   # list of lists = includes time.
        
        points_tups = []
        for hr_of_points in points:
              hr_of_points_tup = [(pt.y, pt.x) for pt in hr_of_points if pt is not None]      # GEOPANDAS, TWITTER STORES IN REVERSE
              points_tups.append(hr_of_points_tup)
        
        print(points_tups[-1])
        hm = HeatMapWithTime(points_tups, radius=15, use_local_extrema=True) # the WithTime feature doesn't work.
        hm.add_to(m) 
        # HeatMap(points_tups[-1], radius=15).add_to(m)
    else:
        points_tups = [(pt.y, pt.x) for pt in points if pt is not None]
        HeatMap(points_tups, radius=15).add_to(m)
        
    m.save(filepath)
    print(f"Heatmap saved to {filepath}")
```

```python
save_heatmap(gdf_mode[c.CENTROID], "user_mode_locs.html")
```

# Loading User Data

```python
def calcRange(num_locs_list):
    '''
    Convert the series for number of unique locations per user to a series representing string range.
    '''
    out = []
    for num_locs in num_locs_list:
        if num_locs == 0:
            out.append("0")
        elif num_locs < 2:
            out.append("1")
        elif num_locs < 5:
            out.append("2-4")
        elif num_locs < 10:
            out.append("5-9")
        elif num_locs < 20:
            out.append("10-19")
        else:
            out.append("20+")
    return out
```

```python
from load_users import load_users_df
```

```python
users_df = load_users_df()
users_df
```

```python
# Track if geotagged.
users_df[c.HAS_TAG] = users_df[c.LOCS].apply(lambda len_geo: len_geo > 0)

# Get birthday -- ISO formats autodetected...
users_df[c.BORN] = pd.to_datetime(users_df[c.BORN])
```

```python
users_df
```

Sort users based on birthday:

```python
geo_users_by_bday_df = users_df[users_df[c.HAS_TAG]==True].groupby([
 users_df[c.BORN].dt.year, 
 users_df[c.BORN].dt.month, 
 users_df[c.BORN].dt.day
]).agg({
  c.U_ID: 'count', 
  c.FLWNG: 'mean',
  c.LOCS: 'mean',
})

users_by_bday_df = users_df.groupby([
 users_df[c.BORN].dt.year, 
 users_df[c.BORN].dt.month, 
 users_df[c.BORN].dt.day
]).agg({
  c.U_ID: 'count', 
  c.FLWNG: 'mean',
  c.LOCS: 'mean',
})

users_by_bday_df
```

```python
geo_users_by_bday_df.index.set_names(['year', 'month', 'day'], inplace=True)
users_by_bday_df.index.set_names(['year', 'month', 'day'], inplace=True)

geo_users_by_bday_df.reset_index(inplace=True)
users_by_bday_df.reset_index(inplace=True)
```

Merge with geo_users:

```python
users_by_bday_df = users_by_bday_df.merge(
  geo_users_by_bday_df, 
  how='left', 
  on=['year','month','day'], 
  suffixes=[None, '_geo']
)
users_by_bday_df
```

Sort out the 19 users "born" in 1970.

```python
users_by_bday_df = users_by_bday_df[users_by_bday_df['year'] > 2000]
users_by_bday_df["U_ID_geo"] = users_by_bday_df["U_ID_geo"].fillna(0)

users_by_bday_df = users_by_bday_df.groupby(["year","month"]).agg({
    'U_ID': 'sum',
    'U_ID_geo': 'sum',
})
```

```python
PERC_GEO = "% with Geotags"

users_by_bday_df[PERC_GEO] = users_by_bday_df["U_ID_geo"].div(users_by_bday_df[c.U_ID])*100
users_by_bday_df
```

```python
##### Range [-0.5, 40km]
ax = users_by_bday_df.plot(use_index=True, y=[c.U_ID])
plt.title("Geotagging by Account Age")

plt.ylabel("Number of Accounts")

users_by_bday_df.plot(ax=ax, use_index=True, y=[PERC_GEO], secondary_y=True)
# plt.xlim(800, 5048)
# plt.ylim(0, 20)
plt.xlabel("Account Birthdate")
plt.ylabel("Percent with Geotag")

plt.show()
```
